```python
text = '''This text contains some emails to send msg as tmr is hloiday enjoy the day.
    G Rakesh CSE 2nd year rakesh@gmail.com
    G Uma EEE 2nd year uma.g@gmrit.edu.in
    G Ramu MECH 4th year ramu@microsoft.com'''
```

```python
doc = nlp(text)
emails = []
for token in doc:
  if token.like_email:
    emails.append(token)
print(emails)
```

```python
nlp = spacy.blank('te')
doc = nlp("హలో రాకేష్, మీరు మైక్రోసాఫ్ట్ ఇన్‌స్టెప్ ఇంటర్న్షిప్‌కు ఎంపికయ్యారు మరియు మీ స్టైపెండ్ నెలకు 50000$ ఉంది. ఒక మంచి రోజు గడపండి మరియు రోజును ఆనందించండి.")
for token in doc:
  print(token, token.is_currency, token.like_num)
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("Gimme", [
    {
        ORTH:"Gim"
    },
    {
        ORTH:"me"
    }
])
doc = nlp("Gimme that book and pen")
tokens = [token.text for token in doc]
print(tokens)
```

```python
doc = nlp("Quantum + AI is the future changing technology and india invests more than $20 billion")
for ent in doc.ents:
  print(ent.text, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```
